```python
message = "Hello My Name Is SARA AHMED and my email is sara.ahmed@university.edu"

print(f"Original message: {message}")

# Q1a: Calculate the length of the message

message_length = len(message)

# Q1b: Convert the message to lowercase

message_lower = message.lower()

# Q1c: Convert the message to uppercase

message_upper = message.upper()

# Q1d: Convert the message to title case

message_title = message.title()

# Q1e: Split the message into individual words

message_words = message.split()

# Q1f: Count how many words are in the message

word_count = len(message_words)

# Print your results

print(f"Message length: {message_length}")

print(f"Lowercase: {message_lower}")

print(f"Uppercase: {message_upper}")

print(f"Title case: {message_title}")

print(f"Word count: {word_count}")
```

```python
# Q2a: Split the message to find the email (hint: look for the word containing @)

email_address = message[-25:]

# Q2b: Check if the email contains both @ and . symbols


# Q2c: Extract just the username part (before the @)

username = email_address[0:10]

# Q2d: Extract just the domain part (after the @)

domain = email_address[-14:]

# Print your results

print(f"Email found: {email_address}")



print(f"Username: {username}")

print(f"Domain: {domain}")
```
